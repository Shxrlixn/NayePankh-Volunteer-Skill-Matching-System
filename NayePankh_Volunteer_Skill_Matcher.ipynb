# NayePankh Volunteer Skill Matching & Recommendation System

## Objective
This project helps NayePankh Foundation assign volunteers to suitable NGO activities based on their skills and interests.

## Features
- Volunteer Database
- Skill-Based Activity Recommendation
- Match Score Calculation
- Volunteer Search
- Skill Filtering
- Analytics Report
- Data Visualization
- CSV Export

```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries Imported Successfully!")
```

```python
volunteers = pd.DataFrame({
    "Name": [
        "Rahul", "Priya", "Amit", "Sneha", "Karan",
        "Anjali", "Rohit", "Pooja", "Vivek", "Meera",
        "Arjun", "Neha", "Sanjay", "Divya", "Akash"
    ],

    "City": [
        "Bangalore", "Mysore", "Bangalore", "Mangalore", "Mysore",
        "Bangalore", "Hubli", "Mysore", "Bangalore", "Mangalore",
        "Hubli", "Bangalore", "Mysore", "Mangalore", "Hubli"
    ],

    "Skills": [
        "Python, AI",
        "Teaching, Communication",
        "Canva, Design",
        "Social Media, Marketing",
        "Python, Teaching",
        "Teaching, Leadership",
        "Marketing, Communication",
        "Canva, Design",
        "AI, Machine Learning",
        "Social Media, Canva",
        "Python, Data Analysis",
        "Communication, Teaching",
        "Event Management",
        "Design, Marketing",
        "AI, Python"
    ]
})

volunteers
```

```python
def recommend_activity(skills):

    skills = skills.lower()

    if "python" in skills or "ai" in skills or "machine learning" in skills:
        return "Technical Mentorship"

    elif "teaching" in skills or "communication" in skills:
        return "Education Program"

    elif "design" in skills or "canva" in skills:
        return "Awareness Campaign"

    elif "marketing" in skills or "social media" in skills:
        return "Outreach Campaign"

    elif "event" in skills:
        return "Volunteer Events"

    else:
        return "General Volunteer Work"


def get_match_score(skills):

    skills = skills.lower()

    if "python" in skills or "ai" in skills:
        return 95

    elif "teaching" in skills or "communication" in skills:
        return 90

    elif "design" in skills or "canva" in skills:
        return 88

    elif "marketing" in skills or "social media" in skills:
        return 85

    elif "event" in skills:
        return 80

    else:
        return 70


print("Recommendation Engine Ready!")
```

```python
volunteers["Recommended Activity"] = volunteers["Skills"].apply(recommend_activity)

volunteers["Match Score"] = volunteers["Skills"].apply(get_match_score)

volunteers
```

```python
search_name = input("Enter Volunteer Name: ")

result = volunteers[
    volunteers["Name"].str.contains(search_name, case=False, na=False)
]

print("\nSearch Result:")
display(result)
```

```python
skill = input("Enter Skill: ")

filtered = volunteers[
    volunteers["Skills"].str.contains(skill, case=False, na=False)
]

print("\nFiltered Volunteers:")
display(filtered)
```

```python
city = input("Enter City: ")

city_filtered = volunteers[
    volunteers["City"].str.contains(city, case=False, na=False)
]

print(f"\nVolunteers from {city}:")
display(city_filtered)
```

```python
top_volunteers = volunteers.sort_values(
    by="Match Score",
    ascending=False
)

print("Top Recommended Volunteers:")
top_volunteers.head(5)
```

```python
print("===== NayePankh Volunteer Analytics Report =====")

print("\nTotal Volunteers:")
print(len(volunteers))

print("\nVolunteers by City:")
print(volunteers["City"].value_counts())

print("\nRecommended Activities:")
print(volunteers["Recommended Activity"].value_counts())

print("\nAverage Match Score:")
print(round(volunteers["Match Score"].mean(), 2))
```

```python
activity_count = volunteers["Recommended Activity"].value_counts()

plt.figure(figsize=(8,5))
activity_count.plot(kind="bar")

plt.title("Volunteer Distribution by Activity")
plt.xlabel("Activity")
plt.ylabel("Number of Volunteers")

plt.show()
```

```python
city_count = volunteers["City"].value_counts()

plt.figure(figsize=(8,5))
city_count.plot(kind="bar")

plt.title("Volunteers by City")
plt.xlabel("City")
plt.ylabel("Number of Volunteers")

plt.show()
```

```python
volunteers.to_csv(
    "NayePankh_Volunteer_Report.csv",
    index=False
)

print("CSV Report Exported Successfully!")
```

```python
highest_score = volunteers["Match Score"].max()

best_volunteers = volunteers[
    volunteers["Match Score"] == highest_score
]

print("Top NGO Volunteer Candidates")
best_volunteers[["Name", "City", "Skills", "Recommended Activity", "Match Score"]]
```

```python
name = input("Enter Volunteer Name: ")
city = input("Enter City: ")
skills = input("Enter Skills: ")

activity = recommend_activity(skills)
score = get_match_score(skills)

new_volunteer = pd.DataFrame({
    "Name": [name],
    "City": [city],
    "Skills": [skills],
    "Recommended Activity": [activity],
    "Match Score": [score]
})

volunteers = pd.concat([volunteers, new_volunteer], ignore_index=True)

print("\nVolunteer Added Successfully!")
display(volunteers.tail())
```

```python
print("===== NayePankh Volunteer Recommendation System =====")

name = input("Enter Volunteer Name: ")
city = input("Enter City: ")
skills = input("Enter Skills: ")

activity = recommend_activity(skills)
score = get_match_score(skills)

print("\n===== Recommendation Result =====")
print(f"Volunteer: {name}")
print(f"City: {city}")
print(f"Skills: {skills}")
print(f"Recommended Activity: {activity}")
print(f"Match Score: {score}%")
```

```python
volunteers.to_csv("NayePankh_Volunteer_Report.csv", index=False)

print("Updated volunteer data saved successfully!")
```
